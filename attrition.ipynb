## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
#Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()


```

```python
#Create a list of at least 10 column names to use as X data
X_columns = ['Age', 'JobSatisfaction', 'OverTime', 'DistanceFromHome', 'Education', 'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked']





#Create X_df using your selected columns
X_df = attrition_df[X_columns]
X_df.head()



#Show the data types for X_df
X_df.dtypes

```

```python
#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)


```

```python
#Convert your X data to numeric data types however you see fit
# Convert categorical features to numeric using one-hot encoding
X_encoded = pd.get_dummies(X_df, columns=['Age', 'JobSatisfaction', 'OverTime', 'DistanceFromHome', 'Education', 'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked'])

# Check data types and convert any remaining non-numeric columns
for column in X_encoded.columns:
    if X_encoded[column].dtype == object:
        try:
            X_encoded[column] = pd.to_numeric(X_encoded[column])
        except ValueError:
            print(f"Column {column} could not be converted to numeric.")

# Display the updated DataFrame
X_encoded.head()


#Determine training data
X_train = pd.get_dummies(X_train)
#Determine testing data
X_test = pd.get_dummies(X_test)


```

```python
#Create a StandardScaler
scaler = StandardScaler()


#Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)


#Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)





```
